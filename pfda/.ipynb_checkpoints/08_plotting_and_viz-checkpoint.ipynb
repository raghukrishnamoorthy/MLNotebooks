```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax1.plot(np.random.randn(50).cumsum(), 'k--')
_ = ax2.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
```

```python
#Creating subplots easily
fig, axes = plt.subplots(2, 3,sharex= True)
axes[0, 1].plot(np.random.randn(10))
axes[1, 2].plot(np.random.randn(10), 'k--')
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins = 50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
plt.plot(np.random.randn(30).cumsum(), 'ko--')
np.random.randn(30).cumsum()
```

```python
plt.plot(np.random.randn(15).cumsum(), color = 'r', linestyle='dashed', marker='o')
```

```python
# Non-linear interpolation
data = np.random.randn(30).cumsum()
plt.plot(data, 'k-', drawstyle='steps-post', label = 'steps-post')
```

###### Ticks, Labels and Legends

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
_ = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                            rotation=30, fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

```python
fig,ax = plt.subplots(1, 1)
ax.plot(np.random.randn(15).cumsum(), 'k', label='one')
ax.plot(np.random.randn(15).cumsum(), 'r--', label='two')
ax.plot(np.random.randn(15).cumsum(), 'b.', label='three')
ax.legend(loc='best')
ax.text(5, 5, 'Hello world')
```
